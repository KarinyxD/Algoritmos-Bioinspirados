```python
import random

def g06(x1, x2):
    return (x1 - 10)**3 + (x2 - 20)**3

def g1(x1, x2):
    return -(x1 - 5)**2 - (x2 - 5)**2 + 100

def g2(x1, x2):
    return (x1 - 6)**2 + (x2 - 5)**2 - 82.81

def main():
    array_minimum = []
    for x in range(1000):
        x1 = random.randint(13, 100)
        x2 = random.randint(0, 100)
        y = g06(x1, x2)
        y_g1 = g1(x1, x2)
        y_g2 = g2(x1, x2)
        
        if y_g1 >= 0:
            y += y_g1 / 100
        if y_g2 >= 0:
            y += y_g2 / 100
        
        array_minimum.append((x1, x2, y))
    
    melhor = array_minimum[0]
    for t in array_minimum:
        if t[2] < melhor[2]:
            melhor = t
    print(melhor)

main()


#restricoes:
# g1: -(x1 - 5)**2 - (x2 - 5)**2 + 100 <= 0
# g2: (x1 - 6)**2 + (x2 − 5)**2 - 82.81 <= 0
```

Função g06: (x1 - 10)³ + (x2 - 20)³
A solução ótima x* = (14.09500000000000064, 0.8429607892154795668) 
f (x∗) = −6961.81387558015.

```python
import random

def g06(x1, x2):
    return (x1 - 10)**3 + (x2 - 20)**3

#restrição 1: -(x1 - 5)² - (x2 - 5)² + 100 <= 0
def g1(x1, x2):
    return -(x1 - 5)**2 - (x2 - 5)**2 + 100

#restrição 2: (x1 - 6)² + (x2 − 5)² - 82.81 <= 0
def g2(x1, x2):
    return (x1 - 6)**2 + (x2 - 5)**2 - 82.81

def main():
    array_minimum = []
    for x in range(10000):
        x1 = random.uniform(13, 100)
        x2 = random.uniform(0, 100)
        y = g06(x1, x2)
        y_g1 = g1(x1, x2)
        y_g2 = g2(x1, x2)
        
        if y_g1 > 0:
            y += y_g1 * 1000
        if y_g2 > 0:
            y += y_g2 * 1000
        
        array_minimum.append((x1, x2, y))
    
    melhor = array_minimum[0]
    for t in array_minimum:
        if t[2] < melhor[2]:
            melhor = t
    print(melhor)

main()
```

Função g24: -x1 - x2
global minimum = x* = (2.32952019747762, 3.17849307411774) 
f (x*) = −5.50801327159536.

```python
import random

def g24(x1, x2):
    return -(x1 - x2)

#restrição 1: -2x1⁴ + 8x1³ - 8x1² + x2 - 2 <= 0
def g1(x1, x2):
    return -2*x1**4 + 8*x1**3 - 8*x1**2 + x2 - 2

#restrição 2: -4x1⁴ + 32x1³ - 88x1² + 96x1 + x2 - 36 <= 0
def g2(x1, x2):
    return -4*x1**4 + 32*x1**3 - 88*x1**2 + 96*x1 + x2 - 36

def main():
    array_minimum = []
    for x in range(10000):
        x1 = random.uniform(0, 3)
        x2 = random.uniform(0, 4)
        y = g24(x1, x2)
        y_g1 = g1(x1, x2)
        y_g2 = g2(x1, x2)
        
        if y_g1 > 0:
            y += y_g1 * 1000000
        if y_g2 > 0:
            y += y_g2 * 1000000
        
        array_minimum.append((x1, x2, y))
    
    melhor = array_minimum[0]
    for t in array_minimum:
        if t[2] < melhor[2]:
            melhor = t
    print(melhor)

main()
```

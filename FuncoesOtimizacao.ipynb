```python
def sphere(x): #array de variaveis (x, y, z, ...)
    soma = 0
    for i in range(len(x)):
        soma = soma + x[i]**2
    return soma

# print (sphere([0, 0]))

# restricoes y <= 2 e x^2 + y^2 <= 25
def restricao_sphere(ponto):
    x, y = ponto
    if (y <= 2) and (x**2 + y**2 <= 25):
        return sphere(ponto)
    else:
        return "valores nao atendem as restriçoes"

restricao([0, 2])
```

```python
def booth(ponto):
    x, y = ponto
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2

# print (booth([1, 3])) # global minimum

# restricoes 10- <= x e y <= 10
def restricao_booth(ponto):
    x, y = ponto
    if (-10 <= x <= 10) and (-10 <= y <= 10) :
        return booth(ponto)
    else:
        return "valores nao atendem as restriçoes"

restricao_booth([-2, 11])
```

```python
def himmelblau(ponto):
    x, y = ponto
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

# print (himmelblau([3, 2])) # minimum

# restricoes 5- <= x e y <= 5
def restricao_himmelblau(ponto):
    x, y = ponto
    if (-5 <= x <= 5) and (-5 <= y <= 5) :
        return himmelblau(ponto)
    else:
        return "valores nao atendem as restriçoes"

restricao_himmelblau([-2, 2])
```

```python
def three_hump_camel(ponto):
    x, y = ponto
    return 2*x**2 - 1.05*x**4 + (x**6)/6 + x*y + y**2

# print (three_hump_camel([0, 0])) # minimum
    
# restricoes 5- <= x e y <= 5
def restricao_camel(ponto):
    x, y = ponto 
    if (-5 <= x <= 5) and (-5 <= y <= 5):
        return three_hump_camel(ponto)
    else:
        return "valores nao atendem as restriçoes"

restricao_camel([-2, 0])
```
